```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.tools.plotting import table


%matplotlib inline
```

```python
df = pd.read_csv('riders.csv')
df
```

```python
team_by_cost = df.groupby('team').sum().sort_values(by='cost', ascending=False)
team_by_cost
```

```python
def plot_table(df):
    ax = plt.subplot(111, frame_on=False) # no visible frame
    ax.xaxis.set_visible(False)  # hide the x axis
    ax.yaxis.set_visible(False)  # hide the y axis

    table(ax, df)  # where df is your data frame

    plt.savefig('mytable.png')
```

```python
sns.boxplot(data=df, y='team', x='cost', orient='h')
```

```python
df.groupby('team').count().sort_values(by='cost', ascending=False)
```

```python
df.groupby('team').mean().sort_values(by='cost', ascending=False)
```

```python
df.groupby('team').median().sort_values(by='cost', ascending=False)
```

```python
team_mean = df.groupby('team').mean().sort_values(by='cost', ascending=False)
team_total = df.groupby('team').sum().sort_values(by='cost', ascending=False)
```

```python
team_captains = df.groupby('team')['cost'].apply(lambda grp: grp.nlargest(3).sum()).sort_values(ascending=False)
#df.groupby(['team'])['cost'].max()
#df[idx]
plot_table(team_captains)
#team_captains
```
